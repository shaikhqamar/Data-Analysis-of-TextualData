# Data Extraction

### Creating text file naming 'URL_ID.txt'

```python
with open("URL_ID.txt","w+",encoding='utf-8') as url_id:
    textual_data = []
    Response = []
    
    url_index = 0
    for link in df['URL']:
        #print(link)
        url = requests.get(link)
        if url.status_code == 200:
            Response.append(url.status_code)
            print(url_index,' > ',url.status_code,'\n',link)
            soup=bs(url.content,'html.parser')
            
            title=soup.find('h1')# extracting title of website
            #print(title)
            #title=title.text.replace('\n',"  ").replace('/',"")
            title = title.text.strip()
            #print(title)
    
            content=soup.find('div',attrs={'class':'td-post-content'})# extracting only text part
            #content = content.encode('utf-8')
            content=content.text.replace('\xa0',"  ").replace('\n',"  ")# replace end line symbol with space
            txt='.'.join((title,content)) # joiing title and content text
            #print(txt)
            
            url_id.write(txt+'\n')
            textual_data.append(txt)
            url_index+=1
        else:
            Response.append(url.status_code)
            print(url_index,' > ',url.status_code,'\n',link)
            url_index+=1
```

```python
# for filteration of MasterDictionary and Textual by comparing with StopWordList
def excluder(file,to_exclude):
    tmp = []
    for i in file:
        if i.lower() not in to_exclude:
            tmp.append(i)
    return tmp
                
# For fetching the items of StopWordlist and MAsterDictionary
def fetch_items(file):
    #if 'StopWords_Currencies' in a:
    with open(file,encoding="ISO-8859-1")as f:
        word = [i.strip().lower() for i in f]
    return word
```

### 2.2 Percentage of Complex Words

###  Getiing complex words

```python
#data['TextualData'][0]
def get_complex_words(row):
    complex_words = []
    #endwith = []
    vowels = list('aeiouAEIOU')
    tokenize_text = wt(re.sub(r'\W',' ',row))
    for word in tokenize_text:
        vowel_counter = 0
        if word.endswith('es')| word.endswith('ed'):
            #endwith.append(word)
            continue
        else:
            for char in word:
                if char in vowels:
                    vowel_counter+=1
        if vowel_counter>2:
            complex_words.append(word)
    return complex_words
```

```python
def percent_of_complex_words(page):
    complex_words = get_complex_words(page)
    tokenize_text = wt(re.sub(r'\W',' ',page))
    return len(tokenize_text)/len(complex_words)
    
    
```
